Letâ€™s consider a hypothetical microarray experiment, in which RNA isolated from tissue is hybridized to probes for 50 marker genes for liver cells that are fixed in a grid in small microscopic spots. We find that 5 test samples of our liver genes are up-regulated, or more highly expressed compared to the reference sample.

```python
liver_genes = ["ALB","Got1","Gpt","Gsta2","Cdo1","Csad","Nr0b2","Hsd17b9","Hsd17b2","Ste","Ces2","Hpcl2","Ces3","Hao2","Fabp7","Amacr","RoDHII","Rdh3","Gck","Pck1","Pc","Slc37a4","Rnase4","Enpp2","Cyp3a13","Aldh1a1","Cyp1a2","Cyp2c","Fmo1","Cyp17a1","Cyp3a11","Adh4","Bdh","Cml1","Lbp","Igfbp2","Inhbe","Ghr","Igfbp1","Dio1","Afp","Obp3 ","Obp3","Mt1a","Lcn2","Ela1","Es2","Spin2a","Krt2-8","Krt1-18","Rgn","Slc17a1","Aqp7","Abcb1","Slc10a1","Gulo","Rph3al","Ca3"]
```

Without specifying a background gene list, an enrichment analysis of these 5 genes against a cell type database results in significant P-values for several different types of liver cells:

```python
gget.enrichr(liver_genes[:5], database="celltypes", plot=True)
```

This makes sense, since our 10 genes are liver markers after all. However, in the context of the 58 total liver genes tested, only 10 liver genes coming back positive does not seem like much. Indeed, when adding the correct background gene list for this experiment, the enrichment results come back not significant:

```python
gget.enrichr(liver_genes[:5], database="celltypes", background_list=liver_genes[:50], plot=True)
```
